## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df = application_df.drop(columns=['EIN', 'NAME'])
application_df.head()
```

```python
# Determine the number of unique values in each column.
application_df.nunique()
```

```python
# Look at APPLICATION_TYPE value counts for binning
app_types = application_df['APPLICATION_TYPE'].value_counts()
app_types
```

```python
# Scratch work
app_types.iloc[8:].index
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`

application_types_to_replace = list(app_types.iloc[8:].index)

# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
classification_counts = application_df['CLASSIFICATION'].value_counts()
classification_counts
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
classification_clean = classification_counts[classification_counts > 1]
classification_clean
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`

classifications_to_replace = list(classification_counts.iloc[5:].index)

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# List of categories
categorical_list = application_df.select_dtypes(include='object').columns.tolist()
categorical_list
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
app_dummies = pd.get_dummies(application_df[categorical_list])

# Concatenate the new DataFrame with the original
application_df = pd.concat([application_df, app_dummies], axis=1)

# Remove the duplicate, non-encoded columns
application_df = application_df.drop(columns=categorical_list)
application_df.head()
```

```python
# Split our preprocessed data into our features and target arrays
y = application_df['IS_SUCCESSFUL'].values
X = application_df.drop(columns='IS_SUCCESSFUL').values

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, stratify=y)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Hyperparameter Tuning

```python
# Identify the number of features
input_features = X.shape[1]

# Create a function that creates a new Sequential model with hyperparameter options
def create_nn_model(hp):
  # Initialize the Sequential model
  nn_model = tf.keras.models.Sequential()

  # Create a list of activation functions for the tuner to try.
  activation = hp.Choice('activation', ['relu', 'tanh', 'sigmoid'])

  # Allow keras to identify the optimal number of neurons in the first hidden layer.
  nn_model.add(tf.keras.layers.Dense(units=hp.Int('first_units',
                                                  min_value=25,
                                                  max_value=100,
                                                  step=5), activation=activation, input_dim=input_features))

  # Allow keras to identify the optimal number of hidden layers after the first and the number of neurons.
  for i in range(hp.Int('num_layers', 1, 6)):
    nn_model.add(tf.keras.layers.Dense(units=hp.Int('units_' + str(i),
                                                    min_value=25,
                                                    max_value=100,
                                                    step=5), activation=activation))

  # Generic output layer.
  nn_model.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))

  # Compile the model
  nn_model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

  return nn_model
```

```python
# Set up Hyperband optimization with a maximum of 20 epochs and 2 iterations using Keras Tuner.

import keras_tuner as kt

tuner = kt.Hyperband(
    create_nn_model,
    objective='val_accuracy',
    max_epochs=20,
    hyperband_iterations=2)
```

```python
# Run the kerastuner to search for the best hyperparameters.
tuner.search(X_train_scaled, y_train, epochs=20, validation_data=(X_test_scaled, y_test))
```

```python
# Get the best hyperparameters for the model
best_hyperparameter = tuner.get_best_hyperparameters(1)[0]
best_hyperparameter.values
```

```python
# Evaluate best model against full test data
best_model = tuner.get_best_models(1)[0]
model_loss, model_accuracy = best_model.evaluate(X_test_scaled, y_test, verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
